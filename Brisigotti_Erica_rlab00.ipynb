## **A.1** - Fibonacci

Write a function that computes and returns the n-th [Fibonacci number](https://en.wikipedia.org/wiki/Fibonacci_sequence).


```r
# The Fibonacci series is a sequence, starting from 0,1 , in which each number is the sum of the two preceding ones.

# I initialize the index n of the element I want to extract
n = 10
# I define a function to calculate the Fibonacci sequence up until element n
my_Fibonacci <- function(n){
# I initialize the first two numbers
fb_sq = c(0,1)
# I then iterate the addition of the code until the element in position n is filled
for (idx in 1:(n-2)){ fb_sq <- c(fb_sq, fb_sq[idx]+fb_sq[idx+1]) }
# and display such n-th element
tail(fb_sq, n=1)}

sprintf('the n-th element of the Fibonacci sequence %1.0f is',my_Fibonacci(n))
```

## **A.2** - GCD

Compute the greatest common divisor of two integers using [Euclid’s algorithm](https://en.wikipedia.org/wiki/Euclidean_algorithm).

```r
# Euclid algorithm aims to find the greatest commmon divisor by iteratively calculating the reminder of the division 
# between the bigger number and the smaller: at the beginnings such numbers are the input integers, while later they consist
# of the smaller number and the reminder of the previous iteration

# I initialize the two integers of which we want to compute the greatest common divisor (GCD)
n1 <- 720
n2 <- 2800

# I then define a function to calculate the GCD
my_GCD <- function(n1,n2){
# I implement Euclid's algorithm: at every iteration, I calculate the reminder of the divisoin bewteen the bigger integer
# and the smaller integer
a = max(n1,n2)
b = min(n1,n2)
r = b
# the iteration continues until reminder equal to zero is reached
while ( tail( r,n=1 )>0 ){ temp = max(a,r)%%b ; r <- c(r,temp) ; a=max(temp,b) ; b=min(temp,b) }
# the GCD is the penultimate element of the sequence in output of the loop: 
GCD = head( tail(r, n=2), n=1 )
return(GCD)}

sprintf('the greatest common divisor of %i and %i is %i',n1,n2,my_GCD(n1,n2))
```

## **A.3** - Integer decomposition

Given a positive integer number, decompose it, with a list of products of prime numbers. Use one of the algorthms described [here](https://en.wikipedia.org/wiki/Integer_factorization).

```r
# I initialize the positive integer number
n = 420
# and define a function to factorize in prime numbers
my_decomposition <- function(n){
# I initialize a vector where I will store the factors
factors = c()
# and start iterating from the first prime number
i = 2
while(n >= i){
# at every iteration, I check if i is a divisor: in such a case, I update the initial number to the quotient of the division
    if( (n %% i) ==0 ) {factors <- c(factors, i); n <- n/i ; i <- i - 1 ; }
# since the quotient might still be divisble by the same number, I counteract the general increasing of the index that
# characterizes the while loop
    i <- i + 1 }
    return(factors) } 

sprintf('the integer decomposition of %i is',n) ; my_decomposition(n)
```

## **B.2** - Athletic world's races

- The following Web server [2] contains an updated list of all the best timings in official 100 mathletics races.
- Open the results on the Web, reported as a table and create a tibble that contains all the results
- To parse the Web data, use the rvest and tidyverse packages and the following code:

```r
library(rvest)
library(tidyverse)
men100m_html <- read_html("http://www.alltime-athletics.com/m_100ok.htm") 
men100m_html |> html_nodes(xpath="//pre") |> html_text() -> men100m_list
men100m_tbl <- read_fwf (men100m_list) %>% mutate(id = row_number()) ; men100m_tbl
```

-  the resulting tibble has all data represented as ”characters”. Explore the tidyverse packages and manipulate the tibble to add a name of the columns and to convert the columns to the proper format

```r
# firstly, I rename the columns of which I can deduce the meaning
names(men100m_tbl) <- c("x1", "best_time", "x3", "name", "nationality", "date_birth", "x7", "where", "when", "id")
men100m_tbl
```

```r
# since all the columns are already of character type, I just change the ones that aren't characters
men100m_tbl %>% mutate( best_time=as.numeric(best_time), x3=as.numeric(x3), name=as.factor(name), 
                       nationality=as.factor(nationality), date_birth=as.Date(date_birth, "%d.%m.%y"),
                       x7=as.factor(x7), when=as.Date(when, "%d.%m.%y")) -> men100m_tbl
men100m_tbl
```

-  once the men100m tbl is properly formatted, using the ggplot2 package, realize the following plots:
   - evolution of the fastest time as a function of date of the races
   - which country had the most fastest runners for each year?

```r
g <- ggplot(men100m_tbl, aes(when, best_time)) + ylim(c(9.5, 10.1))
options(repr.plot.width=15, repr.plot.height=5)
gg <- g + geom_point(aes(col=nationality)) + labs(y="best_time", x="date", title='men 100m', caption="Source: midwest")
plot(gg)
```
